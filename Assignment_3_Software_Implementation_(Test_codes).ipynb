```python
import datetime

class Employee:
    def __init__(self, name, age, dob, passport):
        self.name = name
        self.age = age
        self.dob = dob
        self.passport = passport

class Manager(Employee):
    def __init__(self, name, age, dob, passport, salespersons):
        super().__init__(name, age, dob, passport)
        self.salespersons = salespersons

    def add_salesperson(self, salesperson):
        self.salespersons.append(salesperson)

class Salesperson(Employee):
    def __init__(self, name, age, dob, passport, manager):
        super().__init__(name, age, dob, passport)
        self.manager = manager

class Car:
    def __init__(self, make, model, price):
        self.make = make
        self.model = model
        self.price = price

class Sale:
    def __init__(self, car, salesperson):
        self.car = car
        self.salesperson = salesperson
        self.manager = salesperson.manager
        self.profit = self.calculate_profit()

    def calculate_profit(self):
        profit = self.car.price - (self.car.price * 0.15)
        salesperson_share = profit * 0.065
        manager_share = profit * 0.035
        company_share = profit - salesperson_share - manager_share
        self.salesperson_share = salesperson_share
        self.manager_share = manager_share
        self.company_share = company_share
        return profit

class Company:
    def __init__(self, name):
        self.name = name
        self.sales = []
        self.employees = []

    def add_employee(self, employee):
        self.employees.append(employee)

    def add_sale(self, sale):
        self.sales.append(sale)

    def get_sales_profit(self):
        total_profit = 0
        for sale in self.sales:
            total_profit += sale.profit
        return total_profit

    def get_employee_sales(self, employee):
        employee_sales = []
        for sale in self.sales:
            if sale.salesperson == employee:
                employee_sales.append(sale)
        return employee_sales

    def get_employee_profit_share(self, employee):
        employee_sales = self.get_employee_sales(employee)
        total_profit_share = 0
        for sale in employee_sales:
            total_profit_share += sale.salesperson_share
        if isinstance(employee, Manager):
            for salesperson in employee.salespersons:
                if salesperson in self.employees:
                    total_profit_share += self.get_employee_profit_share(salesperson) * 0.35
        return total_profit_share

    def get_employee_profit(self, employee):
        total_profit = 0
        employee_sales = self.get_employee_sales(employee)
        for sale in employee_sales:
            total_profit += sale.salesperson_share
        return total_profit

    def get_manager_profit(self, manager):
        total_profit = 0
        manager_sales = self.get_employee_sales(manager)
        for sale in manager_sales:
            total_profit += sale.manager_share
        for salesperson in manager.salespersons:
            if salesperson in self.employees:
                total_profit += self.get_employee_profit(salesperson) * 0.35
        return total_profit

# create company
c = Company("Car Dealership")

# create employees
m1 = Manager("John Doe", 40, datetime.date(1982, 1, 1), "123456789", [])
m2 = Manager("Jane Smith", 35, datetime.date(1987, 1, 1), "987654321", [])
s1 = Salesperson("Bob Johnson", 25, datetime.date(1998, 1, 1), "111111111", [])
s2 = Salesperson("Mary Brown", 30, datetime.date(1993, 1, 1), "999999999", [])

#assign salespersons to managers
m1.add_salesperson(s1)
m2.add_salesperson(s2)

#create cars
car1 = Car("Toyota", "Camry", 2018, 15000, 20000)
car2 = Car("Honda", "Civic", 2020, 20000, 25000)
car3 = Car("Nissan", "Altima", 2017, 12000, 18000)

#add cars to the company inventory
c.add_car(car1)
c.add_car(car2)
c.add_car(car3)

#make sales
sale1 = Sale(car1, s1, m1, datetime.date(2023, 5, 3))
sale2 = Sale(car2, s2, m2, datetime.date(2023, 5, 4))

#print profit
print("Salesperson profit:", sale1.salesperson_profit())
print("Manager profit:", sale1.manager_profit())
print("Company profit:", sale1.company_profit())
```
