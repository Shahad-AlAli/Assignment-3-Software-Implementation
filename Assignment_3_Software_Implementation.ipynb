```python
import datetime   # Import the datetime module for working with dates
import pickle     # Import the pickle module for serializing and deserializing Python objects
import tkinter as tk  # Import the tkinter module for building GUI applications

class Employee: # Defining a class named employee to store information about people how work in the company
    def __init__(self, id, name, DOB, passport, jobTitle, basicSalary): # Defining the constructers for the employee class
        self.id = id # Initializeing the attribute id with the value passed as an argument
        self.name = name # Initializeing the attribute name with the value passed as an argument
        self.DOB = DOB # Initializeing the attribute date of birth(DOB) with the value passed as an argument
        self.passport = passport # Initializeing the attribute passport with the value passed as an argument
        self.jobTitle = jobTitle # Initializeing the attribute job title(jobTitle) with the value passed as an argument
        self.basicSalary = basicSalary # Initializeing the attribute basic salary(basicSalary) with the value passed as an argument
    
    def __str__(self): # Define a method to return a string representation of the Employee object
        return f"ID: {self.id}\nName: {self.name}\nDate of Birth: {self.DOB}\nPassport: {self.passport}\nJob Title: {self.jobTitle}\nBasic Salary: {self.basicSalary}"
    
class Manager(Employee): # Define a class named Manager that inherits from the Employee class
    def __init__(self, id, name, DOB, passport, jobTitle, basicSalary): # Defining the constructers for the manager class
        super().__init__(id, name, DOB, passport, jobTitle, basicSalary) # Call the constructor of the parent class to initialize its attributes
        self.salespersons = [] # Initialize the salespersons attribute with an empty list
        
    def add_salesperson(self, salesperson): # Defining a method to add a salesperson to the list of salespersons for the manager
        self.salespersons.append(salesperson) # Appending the sales person to the list
        
    def remove_salesperson(self, salesperson): # Defining a method to remove a salesperson from the list of salespersons for the manager
        self.salespersons.remove(salesperson) # Removing the sales person from the list
        
    def modify_salesperson(self, salesperson): # Defining a method to modify a salesperson in the list of salespersons for the manager
        index = self.salespersons.index(salesperson) 
        self.salespersons[index] = salesperson

    def get_salespersons(self): # Defining a method to return a list of salespeople to the manager.
        return self.salespersons # Returns all of the information about the sales person to the manager 

    def get_sales(self): # Defining a method that gets the total sales of the manager 
        return None  # The manager does not sale cars 
        
class Salesperson(Employee): # Defining a class named Salesperson that inherits from the Employee class
    def __init__(self, id, name, DOB, passport, jobTitle, basicSalary): # Defining the constructers for the sales person class
        super().__init__(id, name, DOB, passport, jobTitle, basicSalary) # Call the constructor of the parent class to initialize its attributes
        self.sales = [] # Initialize the sales attribute with an empty list
        
    def get_details(self): # Defining a method to return a string representation of the Salesperson object
        return f"{self.name} ({self.id}), {self.DOB}, {self.jobTitle}, {self.basicSalary}, Total Sales: {sum(self.sales)}, Number of Sales: {len(self.sales)}"

    def make_sale(self, car, amount): # Defining a method to add a new sale to the list of sales for the manager
      sale = Sale(car, amount) # Create a new Sale object with given car and amount
      self.sales.append(sale) # Append the sale to the list of sales for the manager
    
    def add_commission(self, month, year, commission): # Defining a method to add a commission to the list of sales for the manager
        self.sales.append((None, commission, (month, year))) # Append the commission to the list of sales for the manager with None for car and a tuple for date
    
    def add_sale(self, sale): # Defining a method to add a sale to the list of sales for the manager
        self.sales.append(sale) # Append the given sale object to the list of sales for the manager

    def get_sales(self): # Defining a method to return the list of sales for the manager
        return self.sales # Return the list of sales for the manager

    def get_sales_profit(self): # Defining a method to return the total profit from all sales for the manager
        return [sale.amount * 0.03 for sale in self.sales] # Calculate the profit for each sale (3% of amount), and return as a list

    def get_commission(self, month, year): # Defining a method to return the commission earned in a given month and year for the manage
        sales_in_month = [sale for sale in self.get_sales() if sale.date.month == month and sale.date.year == year] # Filter the sales list to only include sales from the given month and year
        total_sales_in_month = sum(sale.amount for sale in sales_in_month) # Calculate the total amount of sales in the given month and year
        return total_sales_in_month * 0.05 # Calculate the commission earned (5% of total sales amount)

    
class Sale: # Defining a class for the sales of each person 
    def __init__(self, car, amount): # Defining the constructers for the sales class
        self.car = car # Initializeing the attribute car with the value passed as an argument
        self.amount = amount # Initializeing the attribute amount with the value passed as an argument
        self.date = datetime.date.today() # Set the date of the sale to today's date
        
class Car: # Defining a class for the cars 
    def __init__(self, idNumber, name, price, carType): # Defining the constructers for the car class
        self.idNumber = idNumber # Initializeing the attribute id number with the value passed as an argument
        self.name = name # Initializeing the attribute name with the value passed as an argument
        self.price = price # Initializeing the attribute price with the value passed as an argument
        self.car_type = carType # Initializeing the attribute car type with the value passed as an argument
    
    def get_details(self): # Defining a method to get the details of a car as a dictionary
        return {'ID Number': self.idNumber,
                'Name': self.name,
                'Price': self.price,
                'Type': self.carType}

class Company: # Defining a class named company to implement the car managment software
    def __init__(self, name): # Defining the constructers for the company class
        self.name = name # Initializeing the attribute name with the value passed as an argument
        self.employees = [] # Initialize the employees attribute with an empty list
        self.cars = [] # Initialize the cars attribute with an empty list
    
    def add_employee(self, employee): # Defining a method to add an employee to the list of employees for the company
        self.employees.append(employee) # Appending the employee to the list
    
    def delete_employee(self, employeeID): # Defining a method to remove a salesperson from the list of salespersons for the manager
        for i, employee in enumerate(self.employees): # It iterates over each employee in the list and compares their id to the employee_id provided as an argument.
            if employee.id == employeeID: # When a match is found, the index of that employee in the list is used to remove the employee from the list using the del keyword.
                del self.employees[i] # The index of that employee is deleted
                break # It then breaks out of the loop to ensure that only one employee is removed, in case there are multiple employees with the same id.
    
    def modify_employee(self, employeeID, newEmployee): # This function modifies an employee object in the list of employees for the company based on their employee ID.
        for i, employee in enumerate(self.employees): # Enumerate through the list of employees and check if the current employee has the same ID as the given employeeID.
            if employee.id == employeeID: # If a match is found
                self.employees[i] = newEmployee # Replace the old employee object with the newEmployee object at index i.
                break # It then breaks out of the loop to ensure that only one employee is modified, in case there are multiple employees with the same id.

    def get_employee_details(self, employeeID): # This function retrieves the details of an employee from the list of employees based on their employee ID.
        for employee in self.employees: # Enumerate through the list of employees and check if the current employee has the same ID as the given employeeID.
            if employee.id == employeeID and not isinstance(employee, Manager): # If a match is found
                return employee.get_details() # It returns the employees details 
        return None # If the employee is a Manager, the function returns None.
    
    def add_car(self, car): # This function addes a car object to the list of cars for the company.
        self.cars.append(car) # Appends a car to the list 
    
    def delete_car(self, carID): # This function deletes a car object from the list of cars for the company based on its car ID.
        for i, car in enumerate(self.cars): # Enumerate through the list of cars and check if the current car has the same ID as the given carID.
            if car.idNumber == carID: # If a match is found
                del self.cars[i] # The car with that ID and index is deleted
                break # It then breaks out of the loop to ensure that only one car is deleted, in case there are multiple cars with the same id.
    
    def modify_car(self, carID, newCar): # This function modifies a car object in the list of cars for the company based on its car ID.
        for i, car in enumerate(self.cars): # Enumerate through the list of cars and check if the current car has the same ID as the given carID.
            if car.idNumber == carID: # If a match is found
                self.cars[i] = newCar # The car with that ID and index is modified
                break # It then breaks out of the loop to ensure that only one car is modified, in case there are multiple cars with the same id.
    
    def get_car_details(self, carID): # This function retrieves the details of a car object from the list of cars for the company based on its car ID.
        for car in self.cars:  # loop through the cars in the list of cars
           if car.idNumber == carID: # check if the current car's idNumber matches the carID
              return car.get_details() # if a match is found, return the details of the car using its get_details method

    def add_sale(self, sale): # This function adds a sale object to the list of sales for the company.
        self.sales.append(sale) # append the sale object to the list of sales

    def get_sales_profit(self): # This function calculates and returns the total sales profit of the company.
        total_profit = 0 # initialize the total profit to 0
        for sale in self.sales: # loop through the sales in the list of sales
            total_profit += sale.profit # add the profit of each sale to the total profit
        return total_profit # return the total profit

    def get_employee_sales(self, employee): # This function retrieves all sales objects made by a specific employee.
        employee_sales = [] # initialize a list to hold the sales made by the employee
        for sale in self.sales: # loop through the sales in the list of sales
            if sale.salesperson == employee: # check if the current sale's salesperson matches the employee
               employee_sales.append(sale) # if a match is found, append the sale to the employee_sales list
        return employee_sales # return the list of sales made by the employee

    def get_employee_profit_share(self, employee): # This function calculates and returns the total profit share earned by a specific employee.
        employee_sales = self.get_employee_sales(employee) # get the sales made by the employee using the get_employee_sales method
        total_profit_share = 0 # initialize the total profit share to 0
        for sale in employee_sales: # loop through the sales made by the employee
            total_profit_share += sale.salesperson_share # add the salesperson_share of each sale to the total profit share
        if isinstance(employee, Manager): # check if the employee is an instance of the Manager class
           for salesperson in employee.salespersons: # loop through the salespersons managed by the employee
               if salesperson in self.employees: # check if the salesperson is an employee of the company
                  total_profit_share += self.get_employee_profit_share(salesperson) * 0.35 # recursively calculate the profit share of the salesperson's sales and add it to the total profit share
        return total_profit_share # return the total profit share earned by the employee

    def get_employee_profit(self, employee): # This function calculates and returns the total profit earned by a specific employee.
        total_profit = 0 # initialize the total profit to 0
        employee_sales = self.get_employee_sales(employee) # get the sales made by the employee using the get_employee_sales method
        for sale in employee_sales: # loop through the sales made by the employee
            total_profit += sale.salesperson_share # add the salesperson_share of each sale to the total profit
        return total_profit # return the total profit earned by the employee

    def get_manager_profit(self, manager): # This function calculates and returns the total profit earned by a specific manager and their sales team.
        total_profit = 0 # initialize the total profit to 0
        manager_sales = self.get_employee_sales(manager) # Get the sales made by the manager using the get_employee_sales method
        for sale in manager_sales: # For each sale made by the manager and their sales team: 
            total_profit += sale.manager_share # Adding the manager's share of profit to the total profit
        for salesperson in manager.salespersons: # For each sales person the in managers list 
            if salesperson in self.employees: # if the salesperson is still employed by the company
                total_profit += self.get_employee_profit(salesperson) * 0.35 # Add 35% of the total profit earned by the salesperson to the total profit earned by the manager and their sales team
        return total_profit # return the total profit earned by the manager and their sales team.
```
